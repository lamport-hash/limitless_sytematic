```python
from pathlib import Path

import pandas as pd
import numpy as np

SOURCE_DIR = Path("/app/data/import/firstrate/etf_1h")
TARGET_BASE = Path("/app/data/normalised/candle_1hour/firstrate/etf")
EXCHANGE_NAME = "firstrate"
FREQ = "1hour"

g_close_col = "S_close_f32"
g_open_col = "S_open_f32"
g_high_col = "S_high_f32"
g_low_col = "S_low_f32"
g_volume_col = "S_volume_f64"
g_open_time_col = "S_open_time_i"
g_close_time_col = "S_close_time_i"


def extract_symbol_from_filename(p_filename: str) -> str:
    name = Path(p_filename).stem
    parts = name.split("_")
    return parts[0]


def normalize_txt_to_parquet(p_txt_path: Path, p_output_path: Path) -> bool:
    try:
        df = pd.read_csv(
            p_txt_path,
            header=None,
            names=["_timestamp_str", "Open", "High", "Low", "Close", "Volume"],
            dtype={
                "_timestamp_str": str,
                "Open": float,
                "High": float,
                "Low": float,
                "Close": float,
                "Volume": float,
            },
        )

        if df.empty:
            print(f"  Empty file: {p_txt_path.name}")
            return False

        df["Datetime"] = pd.to_datetime(df["_timestamp_str"])
        df = df.sort_values("Datetime").reset_index(drop=True)

        df = df.rename(
            columns={
                "Open": g_open_col,
                "High": g_high_col,
                "Low": g_low_col,
                "Close": g_close_col,
                "Volume": g_volume_col,
                "Datetime": g_open_time_col,
            }
        )

        df[g_close_time_col] = df[g_open_time_col]

        df[g_open_col] = df[g_open_col].astype(np.float32)
        df[g_high_col] = df[g_high_col].astype(np.float32)
        df[g_low_col] = df[g_low_col].astype(np.float32)
        df[g_close_col] = df[g_close_col].astype(np.float32)
        df[g_volume_col] = df[g_volume_col].astype(np.float64)

        start_date = df[g_open_time_col].min().strftime("%Y%m%d")
        end_date = df[g_open_time_col].max().strftime("%Y%m%d")

        symbol = extract_symbol_from_filename(p_txt_path.name)
        output_filename = (
            f"{EXCHANGE_NAME}_{symbol}_{start_date}_{end_date}_{FREQ}.df.parquet"
        )
        output_path = p_output_path / output_filename

        p_output_path.mkdir(parents=True, exist_ok=True)

        df[
            [
                g_open_col,
                g_high_col,
                g_low_col,
                g_close_col,
                g_volume_col,
                g_open_time_col,
                g_close_time_col,
            ]
        ].to_parquet(output_path, index=False)
        print(f"  Created: {output_path}")
        return True

    except Exception as e:
        print(f"  Error processing {p_txt_path.name}: {e}")
        return False
```

```python
def main():
    total_processed = 0
    txt_files = sorted(SOURCE_DIR.glob("*.txt"))

    print(f"Found {len(txt_files)} txt files to process")
    print(f"Source: {SOURCE_DIR}")
    print(f"Target: {TARGET_BASE}")
    print("-" * 50)

    for txt_file in txt_files:
        symbol = extract_symbol_from_filename(txt_file.name)
        output_dir = TARGET_BASE / symbol

        if normalize_txt_to_parquet(txt_file, output_dir):
            total_processed += 1

    print("-" * 50)
    print(f"Complete! Processed {total_processed} symbols")
```

```python
main()
```
